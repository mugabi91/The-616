```python
import pandas as pd


path =r"/content/cancer.csv"


df = pd.read_csv(path)
df.head()

from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier as Rfc
from sklearn.tree import DecisiontreeClassifier as Dtc
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report
import numpy as np
from sklearn.preprocessing import StandardScaler
#selecting my inputs and target

X = df.drop('diagnosis(1=m, 0=b)',axis=1)

#target
y = df['diagnosis(1=m, 0=b)']

#spliting and shuffling data set

X_train, X_test, y_train, y_test = train_test_split(X , y ,test_size=0.2 , shuffle=True, stratify=y)

#scaling data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)



results ={}
cross_Val_dict ={}

def fit_model(model_instance, model_name, Xtrain, Xtest, ytrain, ytest):
    model_instance.fit(X_train, y_train)
    y_pred = model_instance.predict(X_test)
    score = accuracy_score(y_test,y_pred)
    results[model_name] = [score]
    crossval = cross_val_score(model_instance,X_train, y_train)
    cross_Val_dict[model_name] = [np.mean(crossval)]

    return results, cross_Val_dict
#instances
LogReg = LogisticRegression(solver='liblinear')
supportVec = SVC()
RandForest = Rfc()
DecisionTree = Dtc()

models= [LogReg,DecisionTree,supportVec, RandForest]
names = ['LogisticRegression','DecisionTree','SupportVector','RandomForest']
for model,name in zip(models,names):
    fit_model(model_instance=model, model_name=name, Xtrain=X_train_scaled, Xtest=X_test_scaled, ytrain=y_train, ytest=X_test)
#creating DAta frame to evaluate  accuracy results
df_0 =pd.DataFrame(results)
df_0 = df_0.T
df_0.rename(columns={0:"Normal_score"},inplace=True)


df_1 =pd.DataFrame(cross_Val_dict)
df_1 = df_1.T
df_1.rename(columns={0:"Average_crossVad_score"},inplace=True)

new_df= pd.concat([df_0,df_1],axis=1)
new_df
for model,name in zip(models,names):
    fit_model(model_instance=model, model_name=name, Xtrain=X_train_scaled, Xtest=X_test_scaled, ytrain=y_train, ytest=X_test)
#creating DAta frame to evaluate  accuracy results
df_0 =pd.DataFrame(results)
df_0 = df_0.T
df_0.rename(columns={0:"Normal_score"},inplace=True)


df_1 =pd.DataFrame(cross_Val_dict)
df_1 = df_1.T
df_1.rename(columns={0:"Average_crossVad_score"},inplace=True)

new_df= pd.concat([df_0,df_1],axis=1)
new_df
```
